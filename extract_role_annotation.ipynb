В этой тетрадке находятся скрипты, которые необходимо исполнить, чтобы извлечь из размеченного в Inception json-файла информацию о ролях и их индексах.

```python
import json


with open('output_annotation.json', encoding='utf-8') as f:
    data = json.loads(f.read())
print(data)
entities = data['%FEATURE_STRUCTURES']

for entity in entities:
    if "sofaString" in entity.keys():
        sofa = entity["sofaString"]
        break

role_entities = [entity for entity in entities if 'role' in entity.keys()]
roles = [entity['role'] for entity in role_entities]
role_targets = [entity['@target'] for entity in role_entities]


target_indices = []
for target in role_targets:
    for entity in entities:
        if entity["%ID"] == target:
            indices = (entity['begin'], entity['end'])
            target_indices.append(indices)
            break

targets = []

for indice in target_indices:
    targets.append(sofa[indice[0]:indice[1]])


target_indice_zip = zip(targets, target_indices)
role_target_dict = list(zip(target_indice_zip, roles))
role_target_dicts = []

for role in role_target_dict:
    substring = role[0][0]
    offsets = role[0][1]
    role_name = role[1]
    print(substring, offsets, role_name)
    role_dict = {'substring': substring, 'span': offsets, 'role': role_name}
    role_target_dicts.append(role_dict)

with open('roles_extracted.json', 'w', encoding='utf-8') as f:
    json.dump(role_target_dicts, f, indent=4, ensure_ascii=False)
```
